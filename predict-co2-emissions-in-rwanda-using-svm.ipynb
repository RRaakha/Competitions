```python
import pandas as pd
train_data = pd.read_csv('/kaggle/input/playground-series-s3e20/train.csv')
test_data = pd.read_csv('/kaggle/input/playground-series-s3e20/test.csv')
train_data.head()
submissionsample = pd.read_csv('/kaggle/input/playground-series-s3e20/sample_submission.csv')
```

```python
x = train_data.drop(columns=['ID_LAT_LON_YEAR_WEEK', 'year', 'emission'],axis = 1).fillna(0)
y = train_data['emission']
test_data = test_data.drop(columns=['ID_LAT_LON_YEAR_WEEK', 'year'], axis=1).fillna(0)
```

```python
from sklearn.svm import SVC
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.svm import LinearSVR
```

```python
from sklearn.svm import SVR
svm_poly_reg = SVR(kernel='poly',
                  degree=2, C=100, epsilon=0.1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size = 0.33, random_state= 1)
```

```python
y_test.shape
```

```python
y_train.shape
```

```python
svm_poly_reg.fit(x_train, y_train)
```

```python
svm_poly_reg.score(x_train,y_train)
```

```python
predicted = svm_poly_reg.predict(test_data)
```

```python
pred = pd.DataFrame(predicted)
pred
```

```python
submission = pd.DataFrame({'ID_LAT_LON_YEAR_WEEK': submissionsample.ID_LAT_LON_YEAR_WEEK, 'emission': pred[0]})

submission.to_csv('submission.csv', index=False)
submission
```
